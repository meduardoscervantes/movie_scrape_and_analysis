```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import time
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get("https://www.whats-on-netflix.com/library/movies/")
```

```python
# Wait until page fully loads to run this code
driver.find_element_by_partial_link_text('full list').click()
    
```

```python
# Wait until page fully loads to run this code
driver.back()
```

```python
driver.fullscreen_window()
```

```python
df = pd.DataFrame(columns=['Unnamed: 0', 'Cover', 'Title Name', 'Language','Released', 'Rating', 'IMDb'])
df
```

```python
dfs = []
```

```python
html = driver.page_source
soup = bs(html, 'html.parser')
dfs.append(pd.read_html(html)[0])
len(dfs)
```

```python
for x in dfs:
    df = pd.concat([df,x], axis=0)
df
```

```python
df = df.drop(columns=['Unnamed: 0', 'Cover'])
df
```

```python
df = df.reset_index()
```

```python
df = df.drop(columns=['index'])
df
```

```python
df.to_csv('resources/netflix_movies.csv', index=False)
```
