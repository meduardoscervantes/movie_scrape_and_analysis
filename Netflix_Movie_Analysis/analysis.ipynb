```python
import pandas as pd
import numpy as np
import random
```

```python
df = pd.read_csv('resources/netflix_movies_final_data.csv')
df.head()
```

```python
df.iloc[random.randint(0,len(df))]
```

```python
actors = []
for x in df['actors'].tolist():
    if not x == np.nan:
        for y in str(x).split(','):
            if y not in actors:
                tActor = y
                if y[0] == " ":
                    tActor = y[1:]
                if y[-1] == " ":
                    tActor = y[:-1]
                actors.append(tActor)
```

```python
len(actors)
```

```python
actors[:5]
```

```python
df_copy = df.fillna("N/A").copy()
actor_counts = [len(df_copy[df_copy['actors'].str.contains(x)]) for x in actors]
actor_counts[:5]
```

```python
actor_df = pd.DataFrame(
    {
        'actor': actors,
        'movie_counts': actor_counts
    }
).drop_duplicates()
actor_df.sort_values(by='movie_counts', ascending=False).head(100)
```
